```python
def sum_list(values):
  total = 0
  for value in values:
    total += value
  return total
```

```python
A = [1,2,3,4]
result = sum_list(A)
print(f"리스트의 합 : {result}")
```

```python
price = {'콩나물 해장국' : 4500,
         '갈비탕' : 9000,
         '돈까스' : 8000
}
price
```

```python
type(price)
```

```python
price['팟타이'] = 7000
```

```python
price
```

```python
print("메뉴와 가격")
for dish, cost in price.items():
  print(f"{dish} : {cost}원")
```

```python
for dish in price:
  price[dish] -= 500

print("업데이트된 메뉴와 가격")
for dish, cost in price.items():
  print(f"{dish} : {cost}원")
```

```python
def reverse(s):
  return s[::-1]

A = "ABCDE"
reverse(A)
```

```python
print("원본 문자열", A)
print("반대 문자열", reverse(A))
```

```python
def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32

celsius_to_fahrenheit(32)
```

```python
def fahrenheit_to_celsius(fahrenheit):
  return (fahrenheit - 32) * 5/9

fahrenheit_to_celsius(89.6)
```
