```python
import matplotlib.pyplot as plt

mu1, sigma1 = 10, 2
mu2, sigma2 = -6, 6

standard_Guass = np.random.randn(10000)
Guass1 = mu1 + sigma1 * np.random.randn(10000)
Guass2 = mu2 + sigma2 * np.random.randn(10000)

plt.figure(figsize =(10, 6))
plt.hist(standard_Guass, bins=50, alpha=0.4)
plt.hist(Guass1, bins=50, alpha=0.4)
plt.hist(Guass2, bins=50, alpha=0.4)

plt.show()
```

```python
import math

def calCircle(r):
  area = math.pi * r * r
  circum = 2 * math.pi * r
  return area, circum

radius = float(input("원의 반지름 입력: "))
(a, c) = calCircle(radius)
print("원의 넓이는 ",+str(a),"이고 원의 둘레는",+str(c),"이다")
```
