```python
def get_sum(start, end):
  s = 0
  for i in range(start, end+1):
    s += i
  return s

print(get_sum(1, 100))
```

```python
def sort_num(n1, n2):
  if n1 < n2:
    return n1, n2
  else:
    return n2, n1

print(sort_num(110, 210))
print(sort_num(2100, 80))
```

```python
def factorial(n):
  if n <= 1:
    return 1
  else:
    return n * factorial(n-1)

print("4! = ", factorial(4))
```
